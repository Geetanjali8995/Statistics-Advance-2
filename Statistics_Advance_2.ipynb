Q.1. What is hypothesis testing in statistics?

Ans:Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps you determine whether there's enough evidence in a sample to support or reject a specific claim about a population parameter.

 Key Concepts:

1.Hypotheses:

- Null Hypothesis (H‚ÇÄ): The default or original assumption (e.g., "There is no difference").

- Alternative Hypothesis (H‚ÇÅ or Ha): What you're trying to prove (e.g., "There is a difference").

2.Test Statistic:

- A value calculated from sample data that‚Äôs used to test the hypothesis (e.g., a t-score or z-score).

3.Significance Level (Œ±):

- The probability of rejecting the null hypothesis when it‚Äôs actually true (commonly set at 0.05 or 5%).

4.p-value:

- The probability of obtaining a result as extreme as (or more extreme than)
the observed result, assuming the null hypothesis is true.

Q.2 What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans:

**Null Hypothesis (H‚ÇÄ):**

The null hypothesis is the default assumption or status quo. It suggests that there is no effect, no difference, or nothing has changed.

It's what you're trying to test against.

It‚Äôs often a statement of equality (e.g., =, ‚â•, ‚â§).

Example:

(This means you assume there's no difference from the expected score.)

**Alternative Hypothesis (H‚ÇÅ or Ha):**

The alternative hypothesis is what you believe might be true instead of the null. It suggests that there is a difference, an effect, or a change.

It‚Äôs what you try to find evidence for.

It's usually a statement of inequality (‚â†, <, >).

Example:
H‚ÇÅ: The average test score is not equal to 70.
(This means you're testing to see if there's a significant change.)

Q.3.What is the significance level in hypothesis testing, and why is it important?

Ans:
The significance level, usually denoted as Œ± (alpha), is the threshold you set to decide whether or not to reject the null hypothesis.

It represents the probability of making a Type I error, which means rejecting the null hypothesis when it‚Äôs actually true (a false positive).

 Why is it Important?

 1.Sets the bar for evidence:

- It defines how much evidence you need to reject H‚ÇÄ.
- Lower Œ± = stronger evidence required.

2.Controls risk:

- It limits how often you‚Äôre willing to make a false alarm (rejecting something that‚Äôs actually true).

3.Helps with decision-making:

- If p-value ‚â§ Œ±, you reject H‚ÇÄ (evidence is strong enough).
- If p-value > Œ±, you do not reject H‚ÇÄ (evidence isn‚Äôt strong enough).

 **Example:**

Suppose you're testing whether a new teaching method improves scores.

- H‚ÇÄ: New method has no effect.

- You set Œ± = 0.05 (5% chance you're wrong if you reject H‚ÇÄ).

- You calculate a p-value of 0.03.

Since 0.03 < 0.05, you reject H‚ÇÄ and conclude the new method likely has an effect ‚Äî with a 5% risk of being wrong.

Q.4.What does a P-value represent in hypothesis testing?

Ans:

The p-value is the probability of getting a result at least as extreme as the one you observed if the null hypothesis (H‚ÇÄ) is true.

In simpler terms:

It tells you how surprising your results are under the assumption that H‚ÇÄ is correct.

P-value:Small (e.g. ‚â§ 0.05)

Interpretation: Your result is unlikely under H‚ÇÄ ‚Üí reject H‚ÇÄ


P-value:Large (e.g. > 0.05)

Interpretation: Your result is likely under H‚ÇÄ ‚Üí do not reject H‚ÇÄ


Example:
You're testing if a coin is fair.
- H‚ÇÄ: The coin is fair (P = 0.5).
- You flip it 100 times and get 80 heads.
- The p-value is the probability of getting 80 or more heads out of 100 if the coin is actually fair.

If that p-value is 0.002, it means:

- There‚Äôs a 0.2% chance this result happened by random chance.

- Since 0.002 < 0.05, you‚Äôd likely reject H‚ÇÄ and suspect the coin is not fair.

Q.5.How do you interpret the P-value in hypothesis testing?

Ans:The P-value (probability value) in hypothesis testing helps you determine the strength of evidence against the null hypothesis. Here's how to interpret it:

**Definition:** The P-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true.


**Interpretation:**
- Low P-value (typically ‚â§ 0.05): Strong evidence against the null hypothesis ‚Üí Reject the null hypothesis.

- High P-value (> 0.05): Weak evidence against the null hypothesis ‚Üí Fail to reject the null hypothesis.

Example:
If you're testing whether a new drug is more effective than a placebo:

Null hypothesis (H‚ÇÄ): There is no difference.

If your test yields a P-value of 0.02, it means there's a 2% chance you'd see this result or more extreme if the drug had no real effect. Since 0.02 < 0.05, you'd reject the null and conclude the drug likely works.

Q.14. What is a confidence interval, and how is it used to interpret statistical results?

Ans: A confidence interval (CI) is a range of values that's used to estimate an unknown population parameter (like a mean or proportion) based on sample data. It gives you both an estimate and a measure of uncertainty around that estimate.

**How It's Used to Interpret Results**

1.Estimation: CIs provide a range, not just a point estimate. This helps quantify uncertainty.

2.Hypothesis Testing:

- If a confidence interval for a mean difference (e.g., Drug A ‚àí Drug B) includes 0, it suggests no significant difference.

- If a CI for a proportion includes the value under the null hypothesis, we fail to reject the null.

Interpretation Example:

Let‚Äôs say you compute a 95% CI for a population mean as [48.2, 51.8]:

- This does not mean there's a 95% chance the true mean is in that interval (that‚Äôs a common mistake).

- It means: If you repeated the study 100 times, about 95 of the CIs would contain the true population mean.

Q.20.What is an ANOVA test, and what are its assumptions?

Ans:The ANOVA test (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.

**Purpose of ANOVA**

To test the hypothesis:

- Null hypothesis (H‚ÇÄ): All group means are equal.

- Alternative hypothesis (H‚ÇÅ): At least one group mean is different.

**Assumptions of ANOVA**

1.Independence of Observations:

- Each sample or group must be randomly and independently drawn.

2.Normality

- The data in each group should be approximately normally distributed.

- Assessed using tests like Shapiro-Wilk or visual methods like Q-Q plots.

3.Homogeneity of Variance (Homoscedasticity)

- All groups should have equal variances.

- Tested using Levene‚Äôs test or Bartlett‚Äôs test.

If these assumptions are violated, ANOVA results may not be reliable. In such cases, consider:

- Non-parametric alternatives (e.g., Kruskal-Wallis test)

- Transforming the data






Q.21.  What are the different types of ANOVA tests?

Ans:There are several types of ANOVA tests, each suited to different experimental designs and data structures. Here‚Äôs a breakdown of the main types of ANOVA:

**1.One-Way ANOVA (Single-Factor ANOVA):**

- Use case: Compare the means of three or more independent groups based on one categorical independent variable (factor).

- Example: Comparing average exam scores across three teaching methods.

- Tests if group means are significantly different.

**2. Two-Way ANOVA (Factorial ANOVA):**

- Use case: Evaluate the effect of two independent variables on a continuous dependent variable.

- Example: Study effects of teaching method and gender on test scores.

- Can test for:

  Main effect of Factor A

  Main effect of Factor B

  Interaction effect between A and B

**3. Repeated Measures ANOVA:**

- Use case: Used when the same subjects are measured multiple times (longitudinal data).

- Example: Measuring blood pressure of patients before, during, and after treatment.

- Accounts for within-subject variability.

**4. Mixed-Design ANOVA (Split-Plot ANOVA):**

- Use case: Combines both between-subject and within-subject factors.

- Example: Measuring test scores over time for students across different schools (between-subject: school, within-subject: time).

**5. Multivariate ANOVA (MANOVA):**

- Use case: Extends ANOVA to handle multiple dependent variables simultaneously.

- Example: Assessing the impact of a training program on productivity and job satisfaction together.
- Captures correlations between dependent variables.

**6. ANCOVA (Analysis of Covariance):**

- Use case: ANOVA with one or more continuous covariates that you want to control for.

- Example: Comparing teaching methods on test scores while controlling for prior knowledge.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results
import numpy as np
from scipy.stats import norm

# Sample data
sample = [72, 68, 75, 70, 69, 74, 73, 71, 68, 76]
sample_mean = np.mean(sample)
sample_size = len(sample)

# Known population values
population_mean = 70
population_std = 5  # œÉ known

# Z-test statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.3f}")
print(f"P-value = {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population means.")
```

```python
#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
sample = [72, 68, 75, 70, 69, 74, 73, 71, 68, 76]
sample_mean = np.mean(sample)
sample_size = len(sample)

# Known population parameters
population_mean = 70
population_std = 5  # œÉ
alpha = 0.05  # Significance level

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Critical z-values for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.3f}")
print(f"P-value = {p_value:.4f}")
print(f"Z-critical (¬±) = ¬±{z_critical:.3f}")

# Decision
if abs(z_score) > z_critical:
    print("Reject the null hypothesis: Significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")

# -------- Visualization --------
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution')

# Shaded rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region (right)')
```

```python
#5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualizes Type I and Type II errors using hypothesis testing.

    Parameters:
    - mu0: Mean under H0
    - mu1: Mean under H1
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for one-tailed Z-test (right-tailed)
    z_alpha = norm.ppf(1 - alpha)
    critical_value = mu0 + z_alpha * se

    # Type II error (beta)
    z_beta = (critical_value - mu1) / se
    beta = norm.cdf(z_beta)
    power = 1 - beta

    # X range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label=f'H‚ÇÄ: Œº = {mu0}', color='blue')
    plt.plot(x, y1, label=f'H‚ÇÅ: Œº = {mu1}', color='green')

    # Type I Error (alpha) area
    plt.fill_between(x, y0, where=(x >= critical_value), color='red', alpha=0.3, label='Type I Error (Œ±)')

    # Type II Error (beta) area
    plt.fill_between(x, y1, where=(x < critical_value), color='orange', alpha=0.3, label='Type II Error (Œ≤)')

    # Critical value line
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    # Labels and legend
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print stats
    print(f"Critical Value: {critical_value:.2f}")
    print(f"Z (Œ±): {z_alpha:.3f}")
    print(f"Z (Œ≤): {z_beta:.3f}")
    print(f"Type I Error (Œ±): {alpha:.3f}")
    print(f"Type II Error (Œ≤): {beta:.3f}")
    print(f"Power of the Test (1 - Œ≤): {power:.3f}")

# Run the function
visualize_type1_type2(mu0=100, mu1=105, sigma=15, n=36, alpha=0.05)



```

```python
#8.  Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def compare_z_and_t_test():
    # -------------------------
    # Step 1: Simulate sample data
    # -------------------------
    np.random.seed(42)             # For reproducibility
    mu_pop = 100                   # Known population mean
    sigma_pop = 15                 # Known population std-dev (for Z-test)
    n = 25                         # Sample size

    # Generate sample from a normal distribution
    sample = np.random.normal(loc=103, scale=sigma_pop, size=n)

    # -------------------------
    # Step 2: Perform Z-test (œÉ known)
    # -------------------------
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - mu_pop) / (sigma_pop / np.sqrt(n))
    p_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed p-value

    # -------------------------
    # Step 3: Perform T-test (œÉ unknown)
    t_stat, p_t = stats.ttest_1samp(sample, mu_pop)

    # -------------------------
    # Step 4: Print results
    print("=== Summary ===")
    print(f"Sample size (n):         {n}")
    print(f"Sample mean:             {sample_mean:.3f}")
    print(f"Sample std. deviation:   {np.std(sample, ddof=1):.3f}\n")

    print("---- Z-test (œÉ known) ----")
    print(f"Z statistic:             {z_stat:.3f}")
    print(f"P-value (two-tailed):    {p_z:.4f}\n")

    print("---- T-test (œÉ unknown) ----")
    print(f"T statistic:             {t_stat:.3f}")
    print(f"P-value (two-tailed):    {p_t:.4f}")

    # -------------------------
    # Step 5: Visual comparison (bar plot)
    # -------------------------
    labels = ["Z-test", "T-test"]
    values = [p_z, p_t]

    plt.figure(figsize=(6, 4))
    plt.bar(labels, values, color=['steelblue', 'seagreen'])
    plt.ylabel("P-value")
    plt.title("Comparison of P-values: Z-test vs T-test")
    plt.ylim(0, 1)
    plt.axhline(0.05, color='red', linestyle='--', label='Significance level (Œ± = 0.05)')
    plt.legend()
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Run the function
compare_z_and_t_test()
```

```python
#9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean using t-distribution.

    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound): Tuple containing the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # standard error
    df = n - 1  # degrees of freedom
    t_crit = t.ppf((1 + confidence) / 2, df)  # two-tailed t critical value

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound
# Sample data
sample = [72, 68, 75, 70, 69, 74, 73, 71, 68, 76]

# 95% Confidence Interval
ci = confidence_interval(sample, confidence=0.95)
print(f"95% Confidence Interval for the mean: ({ci[0]:.2f}, {ci[1]:.2f})")
```

**Significance of Confidence Interval:**

A confidence interval gives a range of values within which we can be confident the true population mean lies, based on our sample:

- A 95% confidence interval means:

  ‚ÄúIf we took 100 random samples and built a CI for each, about 95 of them would contain the true population mean.‚Äù

- It reflects:

  The sample mean

  The sample size

  The variability in the data

  The confidence level (usually 90%, 95%, or 99%)

```python
#Q.10 Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error using the t-distribution.

    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
        margin_of_error: The margin of error for the sample mean
    """
    data = np.array(data)
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    # Degrees of freedom
    df = n - 1

    # Critical value from t-distribution
    t_critical = t.ppf((1 + confidence) / 2, df)

    # Margin of Error
    margin_of_error = t_critical * standard_error

    return margin_of_error, sample_mean

# Example usage
sample_data = [72, 68, 75, 70, 69, 74, 73, 71, 68, 76]
confidence_level = 0.95

me, mean = calculate_margin_of_error(sample_data, confidence=confidence_level)
print(f"Sample Mean: {mean:.2f}")
print(f"Margin of Error at {int(confidence_level*100)}% Confidence: ¬±{me:.2f}")
print(f"Confidence Interval: ({mean - me:.2f}, {mean + me:.2f})")
```

```python
#Q.11  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def bayesian_coin_inference(alpha_prior=2, beta_prior=2, n_trials=20, k_success=14, confidence=0.95):
    """
    Performs Bayesian inference using a Beta prior and Binomial likelihood.

    Parameters:
        alpha_prior (int): Prior successes (shape parameter of Beta)
        beta_prior (int): Prior failures (shape parameter of Beta)
        n_trials (int): Number of coin flips
        k_success (int): Number of heads observed
        confidence (float): Credible interval level (default 0.95)

    Outputs:
        - Prior and posterior plots
        - Posterior mean and credible interval
    """
    # Posterior parameters (Bayes update for Beta-Binomial model)
    alpha_post = alpha_prior + k_success
    beta_post = beta_prior + (n_trials - k_success)

    # Create grid and evaluate distributions
    x = np.linspace(0, 1, 1000)
    prior_pdf = beta.pdf(x, alpha_prior, beta_prior)
    post_pdf = beta.pdf(x, alpha_post, beta_post)

    # Plot Prior and Posterior
    plt.figure(figsize=(8, 5))
    plt.plot(x, prior_pdf, label=f"Prior: Beta({alpha_prior}, {beta_prior})", color='skyblue', linewidth=2)
    plt.plot(x, post_pdf, label=f"Posterior: Beta({alpha_post}, {beta_post})", color='orange', linewidth=2)
    plt.axvline(alpha_post / (alpha_post + beta_post), color='red', linestyle='--', label="Posterior Mean")

    plt.title("Bayesian Update: Coin Flip Bias Estimation")
    plt.xlabel("Probability of Heads (p)")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # Posterior Summary
    posterior_mean = alpha_post / (alpha_post + beta_post)
    lower_bound, upper_bound = beta.ppf([(1 - confidence) / 2, (1 + confidence) / 2], alpha_post, beta_post)

    print("=== Bayesian Inference Results ===")
    print(f"Posterior Mean (expected p):      {posterior_mean:.3f}")
    print(f"{int(confidence * 100)}% Credible Interval:     ({lower_bound:.3f}, {upper_bound:.3f})")
    print(f"Posterior Distribution:           Beta({alpha_post}, {beta_post})")

# Run the program with default settings
bayesian_coin_inference()
```

```python
#.Q.16.  Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    # Step 1: Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Step 2: Determine which variance is larger
    if var1 > var2:
        f_stat = var1 / var2
        dfn = len(sample1) - 1  # degrees of freedom numerator
        dfd = len(sample2) - 1  # degrees of freedom denominator
    else:
        f_stat = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1

    # Step 3: Calculate critical F value and p-value
    p_value = 2 * (1 - f.cdf(f_stat, dfn, dfd))  # two-tailed test

    # Step 4: Print results
    print("=== F-test for Equality of Variances ===")
    print(f"Sample 1 Variance: {var1:.4f}")
    print(f"Sample 2 Variance: {var2:.4f}")
    print(f"F-statistic:       {f_stat:.4f}")
    print(f"P-value:           {p_value:.4f}")

    # Step 5: Interpretation
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis ‚Äì variances are significantly different.")
    else:
        print("Conclusion: Fail to reject the null hypothesis ‚Äì no significant difference in variances.")

# Example usage
np.random.seed(42)
sample_A = np.random.normal(loc=50, scale=5, size=30)
sample_B = np.random.normal(loc=52, scale=10, size=30)

f_test(sample_A, sample_B)
```

```python
#Q.19  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols

def check_anova_assumptions(data, dependent_var, group_var):
    """
    Check ANOVA assumptions: normality, independence (plot), and equal variances.

    Parameters:
    - data: pandas DataFrame containing the data.
    - dependent_var: name of the dependent variable (string).
    - group_var: name of the group/categorical variable (string).
    """

    # 1. Fit ANOVA model
    model = ols(f'{dependent_var} ~ C({group_var})', data=data).fit()
    residuals = model.resid

    print("\n### Normality Check (Shapiro-Wilk Test on Residuals) ###")
    shapiro_stat, shapiro_p = stats.shapiro(residuals)
    print(f"Shapiro-Wilk Test Statistic: {shapiro_stat:.4f}, p-value: {shapiro_p:.4f}")
    if shapiro_p > 0.05:
        print("Residuals appear to be normally distributed.")
    else:
        print("Residuals do not appear to be normally distributed.")

    # 2. Plot residuals to visually check independence & normality
    print("\n### Residual Plots ###")
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Q-Q plot for normality
    sm.qqplot(residuals, line='s', ax=axes[0])
    axes[0].set_title("Q-Q Plot of Residuals")

    # Residuals vs fitted values plot for independence
    fitted = model.fittedvalues
    axes[1].scatter(fitted, residuals, alpha=0.7)
    axes[1].axhline(0, color='red', linestyle='--')
    axes[1].set_xlabel("Fitted values")
    axes[1].set_ylabel("Residuals")
    axes[1].set_title("Residuals vs Fitted")

    plt.tight_layout()
    plt.show()

    # 3. Levene's Test for equal variances
    print("\n### Equal Variance Check (Levene's Test) ###")
    grouped_data = [group[dependent_var].values for name, group in data.groupby(group_var)]
    levene_stat, levene_p = stats.levene(*grouped_data)
    print(f"Levene‚Äôs Test Statistic: {levene_stat:.4f}, p-value: {levene_p:.4f}")
    if levene_p > 0.05:
        print("Variances appear to be equal across groups.")
    else:
        print("Variances do not appear to be equal across groups.")
# Example with sample data
df = pd.DataFrame({
    'score': np.random.normal(loc=70, scale=10, size=90),
    'group': np.repeat(['A', 'B', 'C'], 30)
})

check_anova_assumptions(df, dependent_var='score', group_var='group')
```

```python
#Q.20  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
#results
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

def two_way_anova_with_interaction(data, dependent_var, factor1, factor2):
    """
    Perform two-way ANOVA with interaction and visualize the results.

    Parameters:
    - data: pandas DataFrame
    - dependent_var: string, name of dependent variable
    - factor1: string, name of first independent variable
    - factor2: string, name of second independent variable
    """

    # Fit two-way ANOVA model with interaction
    formula = f'{dependent_var} ~ C({factor1}) + C({factor2}) + C({factor1}):C({factor2})'
    model = ols(formula, data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    print("### Two-Way ANOVA Table ###")
    print(anova_table)

    # Interaction Plot
    print("\n### Interaction Plot ###")
    sns.pointplot(x=factor1, y=dependent_var, hue=factor2, data=data, dodge=True, markers='o', capsize=.1)
    plt.title("Interaction between Factors")
    plt.ylabel(dependent_var)
    plt.grid(True)
    plt.show()
# Sample dataset
np.random.seed(0)
df = pd.DataFrame({
    'Score': np.random.normal(loc=70, scale=10, size=120),
    'Teaching_Method': np.repeat(['A', 'B'], 60),
    'Classroom': np.tile(np.repeat(['X', 'Y', 'Z'], 20), 2)
})

# Perform two-way ANOVA
two_way_anova_with_interaction(df, dependent_var='Score',
                                factor1='Teaching_Method',
                                factor2='Classroom')

```

```python
#Q21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(dfn=5, dfd=20, alpha=0.05):
    """
    Visualize the F-distribution and mark the critical region.

    Parameters:
    - dfn: degrees of freedom for numerator
    - dfd: degrees of freedom for denominator
    - alpha: significance level
    """

    # Define the range for the x-axis
    x = np.linspace(0, 5, 500)
    y = stats.f.pdf(x, dfn, dfd)

    # Critical F-value (right tail)
    f_critical = stats.f.ppf(1 - alpha, dfn, dfd)

    # Plot the F-distribution curve
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, 'b-', label=f'F-distribution\n(df1={dfn}, df2={dfd})')
    plt.fill_between(x, y, 0, where=(x >= f_critical), color='red', alpha=0.3, label=f'Critical Region\nŒ± = {alpha}')
    plt.axvline(f_critical, color='red', linestyle='--', label=f'F-critical = {f_critical:.2f}')

    # Labels and legend
    plt.title("F-Distribution and Critical Region")
    plt.xlabel("F-value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_f_distribution(dfn=5, dfd=20, alpha=0.05)
```

```python
#Q.22  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def one_way_anova_with_boxplot(data, dependent_var, group_var):
    """
    Perform a one-way ANOVA and visualize results with boxplots.

    Parameters:
    - data: pandas DataFrame
    - dependent_var: string, name of dependent variable (numeric)
    - group_var: string, name of group/categorical variable
    """

    # Group data
    groups = [group[dependent_var].values for name, group in data.groupby(group_var)]

    # Perform ANOVA
    f_stat, p_val = stats.f_oneway(*groups)
    print("### One-Way ANOVA Results ###")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"p-value: {p_val:.4f}")

    if p_val < 0.05:
        print("Result: Significant difference between group means (reject H0)")
    else:
        print("Result: No significant difference between group means (fail to reject H0)")

    # Visualization
    print("\n### Boxplot of Groups ###")
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=group_var, y=dependent_var, data=data, palette='Set2')
    plt.title("Group-wise Comparison of Means (Boxplot)")
    plt.grid(True)
    plt.show()
# Generate example data
np.random.seed(1)
df = pd.DataFrame({
    'Score': np.concatenate([
        np.random.normal(loc=70, scale=5, size=30),  # Group A
        np.random.normal(loc=75, scale=5, size=30),  # Group B
        np.random.normal(loc=80, scale=5, size=30)   # Group C
    ]),
    'Group': ['A']*30 + ['B']*30 + ['C']*30
})

# Call the function
one_way_anova_with_boxplot(df, dependent_var='Score', group_var='Group')
```

```python
#Q.24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy import stats

def chi_square_variance_test(sample, sigma2_0, alpha=0.05, alternative="two-sided"):
    """
    Hypothesis test for the population variance œÉ¬≤ using the œá¬≤ distribution.

    H‚ÇÄ : œÉ¬≤ = sigma2_0
    H‚ÇÅ : œÉ¬≤ ‚â† sigma2_0  (two‚Äësided, default)
          or  œÉ¬≤ > sigma2_0  (alternative="greater")
          or  œÉ¬≤ < sigma2_0  (alternative="less")

    Parameters
    ----------
    sample : array‚Äëlike
        The data (assumed independent and approximately normal).
    sigma2_0 : float
        Variance under H‚ÇÄ.
    alpha : float, default 0.05
        Significance level.
    alternative : {"two-sided", "greater", "less"}

    Returns
    -------
    stat : float
        œá¬≤ test statistic.
    p_value : float
        P‚Äëvalue for the specified alternative.
    ci : tuple(float, float)
        (1¬†‚Äì¬†Œ±) confidence interval for œÉ¬≤.
    """
    sample = np.asarray(sample)
    n        = sample.size
    df       = n - 1
    s2       = np.var(sample, ddof=1)
    stat     = df * s2 / sigma2_0           # œá¬≤ statistic

    # p‚Äëvalue
    if alternative == "two-sided":
        p_value = 2 * min(stats.chi2.cdf(stat, df), 1 - stats.chi2.cdf(stat, df))
    elif alternative == "greater":          # H‚ÇÅ: œÉ¬≤ > œÉ¬≤‚ÇÄ (right tail)
        p_value = 1 - stats.chi2.cdf(stat, df)
    else:                                   # H‚ÇÅ: œÉ¬≤ < œÉ¬≤‚ÇÄ (left tail)
        p_value = stats.chi2.cdf(stat, df)

    # (1‚ÄëŒ±) confidence interval for œÉ¬≤
    chi2_lower = stats.chi2.ppf(alpha/2,  df)
    chi2_upper = stats.chi2.ppf(1-alpha/2, df)
    ci_lower   = df * s2 / chi2_upper
    ci_upper   = df * s2 / chi2_lower

    return stat, p_value, (ci_lower, ci_upper)
np.random.seed(1)
sample = np.random.normal(loc=0, scale=3, size=20)   # true œÉ = 3 ‚áí œÉ¬≤ = 9

stat, p, ci = chi_square_variance_test(sample,
                                       sigma2_0=9,
                                       alpha=0.05,
                                       alternative="two-sided")

print(f"Sample size (n):        {len(sample)}")
print(f"Sample variance (s¬≤):   {np.var(sample, ddof=1):.4f}")
print(f"œá¬≤ statistic:           {stat:.4f}")
print(f"P‚Äëvalue:                {p:.4f}")
print(f"95‚ÄØ% CI for œÉ¬≤:         ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
#Q.25  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(count1, nobs1, count2, nobs2, alpha=0.05):
    """
    Perform a two-proportion Z-test.

    H‚ÇÄ: p1 = p2 (no difference in proportions)
    H‚ÇÅ: p1 ‚â† p2 (two-tailed test by default)

    Parameters:
    - count1, count2: number of successes in group 1 and group 2
    - nobs1, nobs2: total observations in group 1 and group 2
    - alpha: significance level

    Returns:
    - z_stat: Z statistic
    - p_value: corresponding p-value
    - conclusion: result interpretation
    """
    count = np.array([count1, count2])
    nobs  = np.array([nobs1, nobs2])

    z_stat, p_value = proportions_ztest(count, nobs)

    print("### Z-Test for Two Proportions ###")
    print(f"Group 1: {count1}/{nobs1} = {count1/nobs1:.3f}")
    print(f"Group 2: {count2}/{nobs2} = {count2/nobs2:.3f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Conclusion: Significant difference in proportions (reject H‚ÇÄ at Œ± = {alpha})")
    else:
        print(f"Conclusion: No significant difference in proportions (fail to reject H‚ÇÄ)")
    # Example: Success/failure comparison in two groups
# Group 1: 45 successes out of 100
# Group 2: 30 successes out of 90

z_test_proportions(count1=45, nobs1=100, count2=30, nobs2=90)

```

```python
#Q.26  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def f_test_variance(data1, data2, alpha=0.05):
    """
    Perform an F-test to compare variances of two datasets.

    H‚ÇÄ: œÉ1¬≤ = œÉ2¬≤ (variances are equal)
    H‚ÇÅ: œÉ1¬≤ ‚â† œÉ2¬≤ (variances are different)

    Parameters:
    - data1, data2: Arrays or lists of numeric values
    - alpha: Significance level (default = 0.05)

    Returns:
    - Prints F-statistic, p-value, and conclusion
    - Plots F-distribution with critical regions and test statistic
    """

    data1 = np.array(data1)
    data2 = np.array(data2)

    # Sample sizes and variances
    n1, n2 = len(data1), len(data2)
    var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)

    # F-statistic: larger variance / smaller variance (to ensure F ‚â• 1)
    if var1 >= var2:
        f_stat = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    # Two-tailed p-value
    p_value = 2 * min(stats.f.cdf(f_stat, dfn, dfd), 1 - stats.f.cdf(f_stat, dfn, dfd))

    # Critical values
    f_crit_low = stats.f.ppf(alpha / 2, dfn, dfd)
    f_crit_high = stats.f.ppf(1 - alpha / 2, dfn, dfd)

    # Print results
    print("### F-Test for Variance Comparison ###")
    print(f"Sample 1 Variance: {var1:.4f}, Sample Size: {n1}")
    print(f"Sample 2 Variance: {var2:.4f}, Sample Size: {n2}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical F range (Œ±={alpha}): < {f_crit_low:.4f} or > {f_crit_high:.4f}")

    if f_stat < f_crit_low or f_stat > f_crit_high:
        print("Conclusion: Variances are significantly different (reject H‚ÇÄ)")
    else:
        print("Conclusion: No significant difference in variances (fail to reject H‚ÇÄ)")

    # Plot F-distribution
    x = np.linspace(0.1, 5, 500)
    y = stats.f.pdf(x, dfn, dfd)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, 'b-', label='F-distribution')
    plt.axvline(f_stat, color='black', linestyle='--', label=f'F-stat = {f_stat:.2f}')
    plt.fill_between(x, y, 0, where=(x < f_crit_low) | (x > f_crit_high), color='red', alpha=0.3, label='Rejection Region')
    plt.axvline(f_crit_low, color='red', linestyle='--', label=f'Critical Low = {f_crit_low:.2f}')
    plt.axvline(f_crit_high, color='red', linestyle='--', label=f'Critical High = {f_crit_high:.2f}')

    plt.title("F-distribution: Variance Comparison")
    plt.xlabel("F-value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()
  # Simulate two samples with different variances
np.random.seed(0)
sample1 = np.random.normal(0, 5, 30)    # variance ‚âà 25
sample2 = np.random.normal(0, 3, 30)    # variance ‚âà 9

f_test_variance(sample1, sample2)

```

```python
#Q.27  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_goodness_of_fit(observed, expected_probs, alpha=0.05):
    """
    Perform Chi-square goodness-of-fit test.

    H‚ÇÄ: Observed frequencies follow the expected distribution.
    H‚ÇÅ: Observed frequencies do not follow the expected distribution.

    Parameters:
    - observed: list or array of observed counts
    - expected_probs: list of expected probabilities (must sum to 1)
    - alpha: significance level (default = 0.05)
    """

    observed = np.array(observed)
    expected = np.array(expected_probs) * sum(observed)

    # Perform chi-square goodness-of-fit test
    chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

    print("### Chi-Square Goodness-of-Fit Test ###")
    print(f"Observed Counts: {observed}")
    print(f"Expected Counts: {np.round(expected, 2)}")
    print(f"Chi-square Statistic: {chi2_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Conclusion: Significant difference (reject H‚ÇÄ at Œ± = {alpha})")
    else:
        print(f"Conclusion: No significant difference (fail to reject H‚ÇÄ)")

    # Bar plot
    categories = [f"C{i+1}" for i in range(len(observed))]
    x = np.arange(len(categories))
    width = 0.35

    plt.figure(figsize=(8, 5))
    plt.bar(x - width/2, observed, width, label='Observed', color='skyblue')
    plt.bar(x + width/2, expected, width, label='Expected', color='lightgreen')
    plt.xticks(x, categories)
    plt.title("Observed vs Expected Frequencies")
    plt.ylabel("Count")
    plt.legend()
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.show()
  # Simulate observed counts (e.g., 4 categories)
np.random.seed(0)
observed_counts = np.random.multinomial(n=100, pvals=[0.25, 0.25, 0.25, 0.25])

# Assume equal expected probabilities
expected_probs = [0.25, 0.25, 0.25, 0.25]

# Run test
chi_square_goodness_of_fit(observed_counts, expected_probs)

```
